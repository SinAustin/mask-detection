```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import cv2
```

## Goal: Detect whether or not a person is wearing a mask in an image and video stream

```python
# Using cv2 haarcascade to detect faces will get the ball rolling fast
face_cascade = cv2.CascadeClassifier('./cascades/haarcascade_frontalface_default.xml')
```

##### Testing the Face Cascade

```python
#reading in the image
test_img = cv2.imread('test images/03.jpg',1)

#converting to grayscale
grey_test_img = cv2.cvtColor(test_img, cv2.IMREAD_GRAYSCALE)

#detecting faces on the grey image using the cascade, this returns the positions of the faces as (x,y,w,h)
faces = face_cascade.detectMultiScale(grey_test_img)
```

```python
plt.imshow(grey_test_img)
```

```python
#output image in color
out_img = cv2.cvtColor(grey_test_img, cv2.COLOR_RGB2BGR)

#showing detected faces(uses face locations from the grey test img atop the colored img)
for (x,y,w,h) in faces:
    cv2.rectangle(out_img,(x,y),(x+w,y+h),(0,255,0),3)
plt.figure(figsize=(10,10))
plt.imshow(out_img)
```
